```python
import matplotlib.pyplot as plt
import numpy as np
import PIL
import os
```

```python
for c in range(299):
    image = PIL.Image.open('./generated_images/synthetic_images_1L/sample_'+str(c)+'.tif')
    frame = np.asarray(image)[:,:,0]

    frame = frame/frame.max()
    for i in range(frame.shape[0]):
        for j in range(frame.shape[1]):
            if(frame[i,j] >= 0):
                frame[i,j] = 1
            else:
                frame[i,j] = 0

    im = PIL.Image.fromarray(frame)
    im.save("/home/dhamodhar/ganum/citygan/citygan/generated_images/binary_synthetic_images_1L/sample_"+str(c)+".tif")
```
