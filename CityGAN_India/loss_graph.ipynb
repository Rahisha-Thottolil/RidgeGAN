```python
import matplotlib.pyplot as plt
import numpy as np
f = open('/home/dhamodhar/ganum/citygan/citygan/out_1k_256/loss.txt')
l = f.readlines()
ld,lg,dx,dgz,dgz2 = [],[],[],[],[]
j = 0
for i in l:
    l2 = i.split(',')
    ld.append(float(l2[0]))
    lg.append(float(l2[1]))
    dx.append(float(l2[2]))
    dgz.append(float(l2[3]))
    dgz2.append(float(l2[4]))
```
